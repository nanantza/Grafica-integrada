```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read excel file from the 'data' subdirectory using a relative path
data = pd.read_excel('../Datos PET/AVE.xlsx', sheet_name=None)

# Concatena DataFrames en uno solo 
df = pd.concat(data.values(), keys=data.keys(), names=["Region"])
df.reset_index(level=0, inplace=True)

# allow plots to appear within the notebook
%matplotlib inline

sns.pairplot(df, x_vars=['DTBZ'], y_vars='RAC',hue="Region", height=7, aspect=0.7,)
```
